```python
import flexmatcher
import os
import pandas as pd
```

```python
path = "../../../res"
schema_path = "../matching"

train_path_1 = os.path.join(path, "cellphones.csv")
train_path_2 = os.path.join(path, "hhm.csv")
test_path_1 = os.path.join(path, "tgdd.csv")

schema_train_path_1 = os.path.join(schema_path, "cellphones.txt")
schema_train_path_2 = os.path.join(schema_path, "hhm.txt")
```

```python
train_schema_match_1 = pd.read_csv(schema_train_path_1, sep=";")
train_schema_match_2 = pd.read_csv(schema_train_path_2, sep=";")

train_schema_match_1
```

```python
train_mapping_1 = {}
for i in range(train_schema_match_1.shape[0]):
    train_mapping_1[train_schema_match_1['cellphones'][i]] = train_schema_match_1['schema'][i]
train_mapping_2 = {}
for i in range(train_schema_match_2.shape[0]):
    train_mapping_2[train_schema_match_2['hhm'][i]] = train_schema_match_2['schema'][i]
```

```python
columns_uses_1 = train_schema_match_1['cellphones'].tolist()
columns_uses_2 = train_schema_match_2['hhm'].tolist()

train_set_1 = pd.read_csv(train_path_1, sep = ";", usecols=columns_uses_1)
train_set_2 = pd.read_csv(train_path_2, sep = ";", usecols=columns_uses_2)
train_set_1 = train_set_1.astype(str)
train_set_2 = train_set_2.astype(str)
train_set_1
```

```python
schema_list = [train_set_1, train_set_2]
mapping_list = [train_mapping_1, train_mapping_2]
```

```python
fm = flexmatcher.FlexMatcher(schema_list, mapping_list, sample_size=100)
fm.train()            # train flexmatcher
```

```python
test_set_1 = pd.read_csv(test_path_1, sep=';')
test_set_1 = test_set_1.astype(str)
```

```python
predicted_mapping = fm.make_prediction(test_set_1)
```
