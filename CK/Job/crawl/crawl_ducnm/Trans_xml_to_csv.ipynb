```python
# đọc link ảnh
from PIL import Image
from urllib.request import urlopen
# import module sys để gọi ra các ngoại lệ
import sys
import os
import numpy as np
# đọc url
import requests
from io import BytesIO
from imutils import paths
```

```python
# đọc dữ liệu xml
import xml.etree.ElementTree as ET
tree = ET.parse('/content/drive/MyDrive/VCCorp/Product_similarity/datafeed.xml')
```

```python
root = tree.getroot()
```

```python
print(root[842][8].text)
```

```python
import pandas as pd
data = {'link': [], 'image_link': [], 'productname': [], 'description': [], 'sale_price': [], 'price': [], 'availability': [], 'condition': [], 'brand': []}
count = 0
for i in range(843):
  # kiểm tra xme có phải điên thoại hay không
  if root[i][0].text.find('dien-thoai') != -1:
    count +=1
    try:
      data['link'].append(root[i][0].text)
      data['image_link'].append(root[i][1].text)
      data['productname'].append(root[i][2].text)
      data['description'].append(root[i][3].text)
      data['sale_price'].append(root[i][4].text)
      data['price'].append(root[i][5].text)
      data['availability'].append(root[i][6].text)
      data['condition'].append(root[i][7].text)
      data['brand'].append(root[i][8].text) 
      
    except:
      print("Có ngoại lệ ",sys.exc_info()[0]," xảy ra tai: ",count)
```

```python
df_data = pd.DataFrame(data)
df_data.to_csv('/content/drive/MyDrive/VCCorp/Product_similarity/data.csv', index=False)  
```
